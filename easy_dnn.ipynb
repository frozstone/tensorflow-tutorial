```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import tensorflow as tf
import numpy as np

tf.logging.set_verbosity(tf.logging.INFO)
```

```python
IRIS_TRAINING = "./dataset/iris_training.csv"
IRIS_TEST = "./dataset/iris_test.csv"
```

```python
tr_set = tf.contrib.learn.datasets.base.load_csv_with_header(filename=IRIS_TRAINING, target_dtype=np.int, features_dtype=np.float32)
ts_set = tf.contrib.learn.datasets.base.load_csv_with_header(filename=IRIS_TEST, target_dtype=np.int, features_dtype=np.float32)
```

```python
val_metrics =  {"accuracy": tf.contrib.learn.MetricSpec(metric_fn=tf.contrib.metrics.streaming_accuracy, prediction_key="classes")}
val_mon = tf.contrib.learn.monitors.ValidationMonitor(ts_set.data, 
                                                      ts_set.target, 
                                                      every_n_steps=50, 
                                                      metrics=val_metrics)

feature_cols = [tf.contrib.layers.real_valued_column("", dimension=4)]
classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_cols, hidden_units=[10,20,10], n_classes=3, model_dir="models/iris_model", config=tf.contrib.learn.RunConfig(save_checkpoints_secs=1))
```

```python
classifier.fit(x=tr_set.data, y=tr_set.target, steps=2000, monitors=[val_mon])
accuracy_score = classifier.evaluate(x=ts_set.data, y=ts_set.target)["accuracy"]
print (accuracy_score)
```

```python
new_samples = np.array([[6.4, 3.2, 4.5, 1.5], [5.8, 3.1, 5.0, 1.7]], dtype=float)
y = classifier.predict(new_samples)
print(list(y))
```
